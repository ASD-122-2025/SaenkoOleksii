# 1 . Виконати теоретичні розрахунки асимптотичної складності алгоритмів сорутвання методами вставки та бульбашки.


1. У найкращому випадку, коли новий елемент вставляється на початок масиву без необхідності перегляду інших елементів, складність становитиме O(n). У найгіршому (і водночас найпоширенішому) випадку маємо складність O(n²).
Часова складність описується формулою:
T(n) = (n - 1)(c₁ + c₂ + c₆) + ((n(n - 1))/2)(c₃ + c₄ + c₅),
де домінуючим членом є n², що і визначає загальну складність.

2. Метод бульбашки:
Цей алгоритм подібний до методу вставки, однак має інший механізм реалізації.
Часова складність описується формулою:
T(n) = (n - 1)c₁ + ((n(n - 1))/2)(c₂ + c₃ + c₄).
У найгіршому і водночас найчастішому випадку його складність також дорівнює O(n²).
