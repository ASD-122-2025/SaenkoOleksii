# 1.–û—Ü—ñ–Ω–∏—Ç–∏ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É –±—ñ–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ—à—É–∫—É –≤ –û-–Ω–æ—Ç–∞—Ü—ñ—ó.
``` python
def bin_search(a_list, x, left, right):
if left > right or len(a_list) == 0:
return -1
middle = (left + right) // 2
if a_list[middle] == x:
return middle
elif (a_list[middle] < x):
return bin_search(a_list, x, middle + 1, right)
else: # a_list[middle] > x
return bin_search(a_list, x, left, middle - 1)
```
–ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É –±—ñ–Ω–∞—Ä–Ω–æ–≥–æ –ø–æ—à—É–∫—É –≤ –û-–Ω–æ—Ç–∞—Ü—ñ—ó —Å–∫–ª–∞–¥–∞—î O(log n), –¥–µ 
ùëõ
n ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —É –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–æ–º—É –º–∞—Å–∏–≤—ñ —á–∏ —Å–ø–∏—Å–∫—É.

# 2. –ù–∞–ø–∏—Å–∞—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—é, —è–∫–∞ –± –º–æ–≥–ª–∞ –ø–æ–≤–µ—Ä—Ç–∞—Ç–∏ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –∑ —Ç–∞–∫–∏–º —Å–∫–ª–∞–¥–æ–º –ø–æ–ª—ñ–≤: (¬´n¬ª, ¬´time¬ª), –¥–µ n ‚Äì —Ä–æ–∑–º—ñ—Ä –º–∞—Å–∏–≤—É –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ—ó linear_searh();
``` python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
def generate_dataframe(sizes, target_value):
    data = []
    for n in sizes:       
        arr = [random.randint(1, 1000) for _ in range(n)] 
        start_time = time.time()   
        linear_search(arr, target_value) 
        end_time = time.time()
        execution_time = end_time - start_time
        data.append([n, execution_time])
    df = pd.DataFrame(data, columns=["n", "time"])
    return df
```


# 3.–ù–∞–ø–∏—Å–∞—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—é, —è–∫–∞ –± –º–æ–≥–ª–∞ –ø–æ–≤–µ—Ä—Ç–∞—Ç–∏ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –∑ —Ç–∞–∫–∏–º —Å–∫–ª–∞–¥–æ–º –ø–æ–ª—ñ–≤: (¬´n¬ª, ¬´time¬ª), –¥–µ n ‚Äì —Ä–æ–∑–º—ñ—Ä –º–∞—Å–∏–≤—É –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ—ó bin_searh();
```python
def bin_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
def create_dataframe(max_n, step):
    data = []
    for n in range(step, max_n + 1, step):
        arr = sorted(random.sample(range(1, 10 * n), n))
        target = random.choice(arr)
        start_time = time.time()
        bin_search(arr, target)
        end_time = time.time()
        data.append({"n": n, "time": end_time - start_time})
    df = pd.DataFrame(data)
    return df
```
