    assert len(p) <= len(t)

        for j in range(0, len(p)):

## 2 . Пояснити асимптотику наївного алгоритму пошуку підрядка.
Найгірший випадок:
У найгіршому випадку алгоритм розглядає кожну можливу позицію в тексті від 0 до t - p, тобто виконує перевірку для t - p + 1 початкових позицій. Для кожної такої позиції він може порівнювати до p символів шаблону. Таким чином, загальна кількість порівнянь у найгіршому випадку наближається до t × p, що дає асимптотичну складність O(t × p).

Середній випадок:
У середньому ситуація краща, ніж у найгіршому, однак все ще залишається близькою до O(t × p), особливо якщо в тексті часто трапляються часткові збіги з початком шаблону.

Найкращий випадок:
Існує два сценарії для найкращого випадку:

Якщо шаблон узагалі не зустрічається в тексті — тоді кожен символ тексту буде перевірено один раз, що дає складність O(t).

Якщо шаблон збігається з самого початку тексту — тоді буде потрібно лише O(p) порівнянь.

## 3. Обчислити асимптотичну складність алгоритму стиснення рядка за допомогою 𝑧-функції.
Цей процес складається з двох основних етапів: перший — це обчислення Z-функції, другий — безпосереднє виконання стиснення. Обчислення Z-функції має складність O(n), де n — довжина вхідного рядка. Аналогічно, етап стиснення також виконується за O(n), оскільки ми один раз проходимо по масиву Z-функції для пошуку повторюваних префіксів. У результаті загальна складність алгоритму стиснення з використанням Z-функції дорівнює O(n) + O(n) = O(n).

## 4. Контрольні питання
'Що таке «префікс-функція» у контексті алгоритмів на рядках, і як вона відрізняється від Z-функції? Префікс-функція показує довжину найбільшого префікса, що є суфіксом підрядка. Z-функція визначає довжину префікса, який збігається з початковою частиною рядка.

Що таке Z-функція у вищому контексті алгоритмів на рядках, і яка її роль у вирішенні задач? Z-функція – це масив, що оцінює збіги з початковим префіксом рядка. Використовується для пошуку підрядків, обчислення унікальних підрядків і стискання рядків.

Які існують підходи до вирішення задачі «найдовший спільний підрядок» для двох рядків? Динамічне програмування, Суфіксні дерева, Бінарний пошук із хешуванням.

Як можна застосувати алгоритми на рядках у задачах обробки природної мови або обробки текстів? Для пошуку шаблонів, аналізу частот, стиснення тексту, створення індексів і виявлення плагіату.
