## 1. Написати функцію pop_n(), що видаляє елементи стека з його початку до номера n включно;
```python
def pop_n(stack, n):
    for _ in range(n + 1):
        if stack:
            stack.pop()
        else:
            break
```

## 2. Оцінити асимптотичну складність (у середньому і в найгіршому разі) процедур search, insert і delete роботи зі стеком.
Search:
У найгіршому випадку складність дорівнює O(n), оскільки потрібно пройти через увесь стек, припускаючи, що потрібний елемент знаходиться в самому кінці.
У середньому випадку також маємо O(n), адже елемент, імовірно, розташований ближче до середини стеку, проте все одно доводиться переглядати послідовно всі попередні елементи.

Insert:
У найгіршому випадку складність становить O(1), оскільки додавання нового елемента на вершину стеку займає сталий час.
У середньому випадку складність також дорівнює O(1), оскільки процес аналогічний найгіршому випадку.

Delete:
У найгіршому випадку складність — O(1), подібно до Insert, але тут ми видаляємо верхній елемент стеку.
У середньому випадку складність також O(1), оскільки процедура видалення завжди займає однаковий час.

## 3. написати функцію print_n(), що друкує елементи черги з його початку до номера n включно;
``` python
def print_n(queue, n):
    for i in range(n + 1):
        if i < len(queue):
            print(queue[i])
        else:
            break
```

## 4. Оцінити асимптотичну складність (у середньому і в найгіршому разі) процедур search, insert і delete роботи з чергою.
Search : Найгірший випадок - O(n) ми проходимо всю чергу бо вважаємо що елемент у кінці в найгіршому випадку.Середній випадок - O(n) тут ми вважаємо що він знаходиться в центрі черги але ми все ще повинні пройти до нього крізь інші елементи черги.

Insert : Найгірший випадок - O(1) ця операція займає сталий час бо ми просто додаємо елемент в кінець черги.Середній випадок - O(1) - аналогічно найгіршому випадку.

Delete : Найгірший випадок - O(1) тут ситуація схожа з Insert але тут ми видаляємо верхній елемент черги.Середній випадок - O(1) - аналогічно найгіршому випадку

## 5. Контрольні питання:
- Стек – це базова структура даних, в якій ми можемо тільки вставляти або видаляти елементи на початку стека.Існує три основні операції, які можуть виконуватися в стеках: вставка елемента в стек (push), видалення елемента з стека (pop) і відображення вмісту стека (pip).
- Основна відмінність між стеком і чергою полягає в принципі, за яким елементи додаються та видаляються , у стеку перший додається та перший виходить , а у черзі додається в кінець та видаляється з верхівки.
-
